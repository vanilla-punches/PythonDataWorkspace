```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
#hue
sns.relplot(x='total_bill', y='tip', data=tips, hue='time')
```

```python
#style
sns.relplot(x='total_bill', y='tip', data=tips, hue='time', style='size')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette='pastel')
```

```python
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
df
```

```python
sns.relplot(x='time', y='value', data=df, palette='pastel', kind='line')
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, ci=None)
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, hue='region', style='event')
```

```python
#cat plot
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex', split=True)
```

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.catplot(x='sex', y='survived', hue='class', data=titanic, kind='bar')
```
