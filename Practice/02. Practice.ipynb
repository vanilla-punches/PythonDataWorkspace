```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
#hue
sns.relplot(x='total_bill', y='tip', data=tips, hue='time')
```

```python
#style
sns.relplot(x='total_bill', y='tip', data=tips, hue='time', style='size')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='time', palette='husl')
```
