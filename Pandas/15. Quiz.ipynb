```python
import pandas as pd
data = {
    'Movies' : ['Interstellar', 'Parasite', 'The Upside', 'Divergent', 'Apocalypto', 'Prometheus', 'The Wolf of Wall Street', 'The Revenant'],
    'Year' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    'Counts' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270],
    'Rating' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

### 1. Extract movies only

```python
df['Movies']
```

### 2. Extract rating from movies

```python
df[['Movies', 'Rating']]
```

### 3. Extract movies and rating that were released after 2015

```python
df.loc[df['Year'] >= 2015, ['Movies', 'Rating']]
```

### 4. Add a column that is described below
"Recommned Rating" = (Counts*Rating)//100

```python
df['Recommend Stars'] = (df['Counts']*df['Rating'])//100
df
```

```python
df.sort_values('Movies', ascending=False)
```
