```python
import pandas as pd
data = {
    'Name' : ['Ian', 'Jason', 'Smith', 'John', 'Kara', 'Daisy', 'Jack', 'Poppy'], 
    'School' : ['A1', 'A2', 'A2', 'A1', 'A1', 'A1', 'A2', 'A1'],
    'Height' : ['178', '198', '168', '188', '201', '172', '174', '180'],
    'English' : ['90', '100', '85', '70', '82', '91', '77', '80'],
    'Math' : ['100', '82', '87', '65', '78', '80', '50', '90'],
    'Biology' : ['88', '78', '90', '95', '88', '76', '85', '70'],
    'Coding' : ['Python', '', 'C++', 'Java', '', '', 'R', 'Python']
}
data

df = pd.DataFrame(data, index=['#1', '#2', '#3', '#4', '#5', '#6', '#7', '#8'])
df
```

```python
df.index
```

## Index name setting

```python
df.index.name = 'Application Number'
df
```

## Index reset

```python
df.reset_index()
```

```python
df.reset_index(drop=True) #Deleting the index you created earlier "Application Number"
# HOWEVER, this only shows how it will looks like when you drop it. No permanent change!
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True) # inplace=True will permanently drop the indext
df
```

## Index setting
Choose a specific column as an index

```python
df.set_index('Name')
```

```python
df
```

```python
df.set_index('Name', inplace=True)
```

```python
df
```

## Index order
ascending and descending order

```python
df.sort_index() # ascending order for index
```

```python
df.sort_index(ascending=False) # descending order
```
