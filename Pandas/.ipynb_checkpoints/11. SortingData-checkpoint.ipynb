# Sorting Data

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='Application Number')
df
```

```python
df.sort_values('Height') # Ascending order
```

```python
df.sort_values('Height', ascending=False)
```

```python
df.sort_values(['Math', 'Finance'], ascending=False)
```

```python
df.sort_values(['Math', 'Finance'], ascending=[True, False]) 
# Math is ascending order
# Finance is descedning order
```

```python
df['Height'].sort_values()
```

```python
df['Height'].sort_values(ascending=False)
```

```python
df.sort_index()
```

```python
df.sort_index(ascending=False)
```
