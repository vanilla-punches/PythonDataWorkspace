# Grouping

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='Application Number')
df
```

```python
df.groupby('School')
```

```python
df.groupby('School').get_group('A1')
```

```python
df.groupby('School').mean()
```

```python
df.groupby('School').size()
```

```python
# group by 'School' then check the size of each school then only show size of A2
df.groupby('School').size()['A2']
```

```python
df.groupby('School')['Height'].mean()
```

```python
df.groupby('School')[['English', 'Math', 'Finance']].mean()
```

```python
df['Nationality'] = ['Korea', 'US', 'Canada', 'Korea', 'Canada', 'US', 'Canada', 'Korea']
df
```

```python
df.groupby(['School', 'Nationality']).mean()
```

```python
df.groupby('Nationality').mean()
```

```python
# Group by 'Nationality' then find the mean of the groups and then sort by 'Math'
df.groupby('Nationality').mean().sort_values('Math')
```

```python
df.groupby(['School', 'Nationality']).sum()
```

```python
df.groupby('School')['Coding'].count()
```

```python
df.groupby('School')[['Name', 'Coding']].count()
```

```python
school = df.groupby('School')
school['Nationality'].value_counts()
```

```python
school['Nationality'].value_counts().loc['A1']
```

```python
# Group by 'School' then get the percent of 'Nationality'
school['Nationality'].value_counts(normalize=True).loc['A1']
```
