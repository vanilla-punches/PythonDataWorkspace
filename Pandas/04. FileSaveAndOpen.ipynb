## 4. File Save and Open
DataFrame ---> excel, csv, txt

```python
import pandas as pd
data = {
    'Name' : ['Ian', 'Jason', 'Smith', 'John', 'Kara', 'Daisy', 'Jack', 'Poppy'], 
    'School' : ['A1', 'A2', 'A2', 'A1', 'A1', 'A1', 'A2', 'A1'],
    'Height' : ['178', '198', '168', '188', '201', '172', '174', '180'],
    'English' : ['90', '100', '85', '70', '82', '91', '77', '80'],
    'Math' : ['100', '82', '87', '65', '78', '80', '50', '90'],
    'Biology' : ['88', '78', '90', '95', '88', '76', '85', '70'],
    'Finance' : ['95', '88', '75', '92', '84', '80', '100', '89'],
    'Coding' : ['Python', '', 'C++', 'Java', '', '', 'R', 'Python']
}
data

df = pd.DataFrame(data, index=['#1', '#2', '#3', '#4', '#5', '#6', '#7', '#8'])
df.index.name = 'Application Number'
df
```

### Save as csv

```python
df.to_csv('score.csv')
```

```python
df.to_csv('score.csv', index = False)
```

### Save as txt

```python
df.to_csv('score.txt', sep='\t')
```

### Save as excel

```python
df.to_excel('score.xlsx')
```

---
## Open
### csv file open

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows=1) # skipping the first row
df
```

```python
df = pd.read_csv('score.csv', skiprows=[2,3,5])
df
```

```python
df = pd.read_csv('score.csv', nrows=4) 
# show first 4 rows, but the first row is considered as title row, 
# but it will show 4 rows after the title row
df
```

```python
df = pd.read_csv('score.csv', skiprows=2, nrows=4) # skip first 2 rows and bring the next 4 rows
df
```

### open txt file

```python
df = pd.read_csv('score.txt', sep='\t')
df
```

```python
df = pd.read_csv('score.txt', sep='\t', index_col='Application Number')
df
```

```python
df = pd.read_csv('score.txt', sep='\t')
df.set_index('Application Number', inplace=True)
df
```

### Open excel file

```python
df = pd.read_excel('score.xlsx')
df
```

```python
df = pd.read_excel('score.xlsx', index_col='Application Number')
df
```
