### Drop na

```python
df = pd.read_excel('score.xlsx', index_col='Application Number')
df
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(inplace=True)
df
```

```python
df = pd.read_excel('score.xlsx', index_col='Application Number')
df
```

- axis: index or columns  
index is about row  
columns is about column. 
- how: any or all  
any will work whenever there is any na  
all will work when there are all na

```python
df.dropna(axis='index', how='any')
```

```python
df.dropna(axis='columns') # how='any' is default
```

```python
df['School'] = np.nan
df
```

```python
df.dropna(axis='columns', how='all')
```
