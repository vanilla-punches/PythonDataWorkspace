# Function

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='Application Number')
df
```

```python
df['School'] = df['School'] + ' University'
df
```

## Function (apply)

```python
def add_cm(Height):
    return str(Height) + 'cm'
df['Height'] = df['Height'].apply(add_cm)
df
```

```python
def capitalize(lang):
    if pd.notnull(lang):
        return lang.capitalize()
    return lang

df['Coding'] = df['Coding'].apply(capitalize)
df
```

```python
# There is an easier way to do it as well. It will give you the same result.
df['Coding'].str.capitalize()
```
