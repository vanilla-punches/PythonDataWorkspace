# 15. Scatter Plot

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Hiragino Sans'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['Levels'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['English'], df['Math'])
plt.xlabel('English Score')
plt.ylabel('Math Score')
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['English'], df['Math'], s=sizes)
plt.xlabel('English Score')
plt.ylabel('Math Score')
```

```python
sizes = df['Levels'] * 500
```

```python
plt.scatter(df['English'], df['Math'], s=sizes)
plt.xlabel('English Score')
plt.ylabel('Math Score')
```

```python
plt.scatter(df['English'], df['Math'], s=sizes, c=df['Levels'], cmap='viridis')
plt.xlabel('English Score')
plt.ylabel('Math Score')
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(df['English'], df['Math'], s=sizes, c=df['Levels'], cmap='viridis')
plt.xlabel('English Score')
plt.ylabel('Math Score')
plt.colorbar()
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(df['English'], df['Math'], s=sizes, c=df['Levels'], cmap='viridis')
plt.xlabel('English Score')
plt.ylabel('Math Score')
plt.colorbar(ticks=[1,2,3], label='Levels')
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(df['English'], df['Math'], s=sizes, c=df['Levels'], cmap='viridis')
plt.xlabel('English Score')
plt.ylabel('Math Score')
plt.colorbar(ticks=[1,2,3], label='Levels', shrink=0.8, orientation='horizontal')
```
