```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Hiragino Sans'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=5)
```

## Marker

```python
plt.plot(x,y, marker='o')
```

```python
plt.plot(x,y, marker='o', linestyle='none')
```

```python
plt.plot(x,y, marker='X', linestyle='none', markersize=10)
```

```python
plt.plot(x,y, marker='X', linestyle='none', markersize=10, markeredgecolor='green')
```

```python
plt.plot(x,y, marker='o', linestyle='none', markersize=20, markeredgecolor='red', markerfacecolor='yellow')
```

## Line style

```python
plt.plot(x,y, linestyle=':')
```

```python
plt.plot(x,y, linestyle='--')
```

```python
plt.plot(x,y, linestyle='-.')
```

## Color

```python
plt.plot(x,y, linestyle='--', color='orange')
```

```python
plt.plot(x,y, color='g')
```

## Format

```python
plt.plot(x,y, 'ro--') # ro-- means color, marker, linestyle
```

```python
plt.plot(x,y, 'bv-.')
```

## Short name

```python
plt.plot(x,y, marker='o', mfc='red', ms=10, mec='blue', ls='--') # short name for marker, line style, etc.
```

## Transparency of lines

```python
plt.plot(x,y, marker='o', mfc='red', ms=10, mec='blue', alpha=0.3)
```

## Graph size

```python
plt.figure(figsize=(10, 5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5, 10), dpi=200) #dpi = dots per inch
plt.plot(x,y)
```

## Background color

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```
