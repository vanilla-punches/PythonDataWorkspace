# Bar Graph

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Hiragino Sans'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['Ian Lee', 'Daft Punk', 'Steven Gerrard']
values = [180, 185, 190] # Height
plt.bar(labels, values)
```

```python
labels = ['Ian Lee', 'Daft Punk', 'Steven Gerrard']
values = [180, 185, 190]
plt.bar(labels, values, color='y')
```

```python
labels = ['Ian Lee', 'Daft Punk', 'Steven Gerrard']
values = [180, 185, 190]
colors = ['g', 'r', 'b'] # Setting indivisual colors
plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
labels = ['Ian Lee', 'Daft Punk', 'Steven Gerrard']
values = [180, 185, 190]
plt.bar(labels, values)
plt.ylim(175, 200) # Setting y-axis range
```

```python
plt.bar(labels, values, width=0.3) # Size of width
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) # Tilting x-axis label
plt.yticks(rotation=45) # Tilting y-axis label
```

```python
labels = ['Ian Lee', 'Daft Punk', 'Steven Gerrard']
values = [180, 185, 190]
ticks = ['1st person', '2nd person', '3rd person']
plt.bar(labels, values)
plt.xticks(labels, ticks)
```
