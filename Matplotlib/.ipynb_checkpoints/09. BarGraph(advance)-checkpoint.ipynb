# 09. Bar Graph(advance)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Hiragino Sans'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['Ian Lee', 'Daft Punk', 'Steven Gerrard']
values = [180, 185, 190]
plt.barh(labels, values) #barh ---> horizontal bar graph
```

```python
plt.barh(labels, values)
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('o') # patterns of bars
bar[1].set_hatch('*')
bar[2].set_hatch('/')
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)
for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center', color='red')
```
