## COVID-19 Vaccination and Types

```python
days = [1, 2, 3] #Jan, Feb, Mar
az = [2, 4, 8] # x100,000 people
pfizer = [5, 1, 3]
moderna = [1, 2, 5]

plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', ls='--')
plt.plot(days, moderna, label='moderna', marker='^', ls='-.')
plt.legend()
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', ls='--')
plt.plot(days, moderna, label='moderna', marker='^', ls='-.')
plt.legend(ncol=3) # Making 3 legends
```
