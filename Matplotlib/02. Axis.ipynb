# 02. Axis

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Hiragino Sans'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # Use this when axis can't show "-" sign
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
plt.title('Line Graph', fontdict={'family':'Sathu', 'size':15})
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
plt.plot(x,y)
plt.xlabel('x-axis', color='red')
plt.ylabel('y-axis', color='blue')
```

```python
plt.plot(x,y)
plt.xlabel('x-axis', color='red', loc='right')
plt.ylabel('y-axis', color='blue', loc='top')
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```
