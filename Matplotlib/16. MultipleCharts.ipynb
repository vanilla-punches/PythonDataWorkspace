# 16. Multiple Charts

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Hiragino Sans'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 15))
fig.suptitle('Summary of Data')

# 1st graph
axs[0, 0].bar(df['Name'], df['English'], label='English Score')
axs[0, 0].set_title('1st Graph')
axs[0, 0].legend()
axs[0, 0].set(xlabel='Name', ylabel='Score')
axs[0, 0].set_facecolor('lightyellow')
axs[0, 0].grid(linestyle='--', linewidth=0.5)

# 2nd graph
axs[0, 1].plot(df['Name'], df['Math'], label='Math')
axs[0, 1].plot(df['Name'], df['Finance'], label='Finance')
axs[0, 1].set_title('2nd Graph')
axs[0, 1].legend()

# 3rd graph
axs[1, 0].barh(df['Name'], df['Height'])
axs[1, 0].set_title('3nd Graph')

# 4th graph
axs[1, 1].plot(df['Name'], df['Biology'], color='green', alpha=0.3)
axs[1, 1].set_title('4th Graph')
```
