# 17. Matplotlib Quiz

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

data = {
    'Movies' : ['Interstellar', 'Parasite', 'The Upside', 'Divergent', 'Apocalypto', 'Prometheus', 'The Wolf of Wall Street', 'The Revenant'],
    'Year' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    'Counts' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270],
    'Rating' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
plt.bar(df['Movies'], df['Rating'])
```

```python
plt.bar(df['Movies'], df['Rating'])
plt.title('Top 8 Movies and Rating')
plt.xlabel('Movies')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
df_group = df.groupby('Year').mean()
df_group
```

```python
plt.plot(df_group.index, df_group['Rating'])
plt.show()
```

```python
# marker is o
# range of x-axis is from 2005 to 2020
# range of y-axis is from 7 to 10
plt.plot(df_group.index, df_group['Rating'], marker='o')
plt.xticks([2005, 2010, 2015, 2020])
plt.ylim(7, 10)
plt.show()
```

```python
filt = df['Rating'] >= 9.0
values = [len(df[filt]), len(df[~filt])]
labels = ['Greater than 9.0', 'Lower than 9.0']

plt.pie(values, labels=labels, autopct='%.1f%%')
plt.legend(loc=(1, 0.3))
plt.show()
```
