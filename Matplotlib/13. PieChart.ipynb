# 13. Pie Chart

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Hiragino Sans'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Jave', 'R', 'SAS', 'C/C++', 'MATLAB']
plt.pie(values, labels=labels)
plt.show()
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90)
plt.show()
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Jave', 'R', 'SAS', 'C/C++', 'MATLAB']
# explode = [0.2, 0.1, 0, 0, 0, 0]
explode = [0.1] * 6
plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.5, 0.3))
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.5, 0.3), title='Languages')
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.title('Languages')
plt.legend(loc=(1.5, 0.3), title='Languages')
plt.show()
```
