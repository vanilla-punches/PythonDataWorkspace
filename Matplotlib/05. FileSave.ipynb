# 0.5 File save

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Hiragino Sans'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
plt.savefig('graph.png')
```

```python
plt.plot(x,y)
plt.savefig('graph.png', dpi=100)
```

```python
plt.figure(dpi=200)
plt.plot(x, y)
plt.savefig('graph_200.png')
```

```python
plt.figure(dpi=200) # figure you see on jupyter
plt.plot(x, y)
plt.savefig('graph_200.png', dpi=100) # figure you see is dpi=200, but saved file's dpi=100
```
