# 14. Pie Chart(advance)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Hiragino Sans'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Jave', 'R', 'SAS', 'C/C++', 'MATLAB']
colors = ['lightcyan', 'violet', 'lightblue', 'tomato', 'orange', 'lavender']
explode = [0.1] * 6

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, colors=colors, explode=explode)
plt.show()
```

```python
wedgeprops={'width':0.6}
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, colors=colors, explode=explode, wedgeprops=wedgeprops)
plt.show()
```

```python
wedgeprops={'width':0.6, 'edgecolor':'w','linewidth':3} # Adding white lines for each language without using explode
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

```python
def custom_autopct(pct):
    return ('%.1f%%' % pct) if pct >= 10 else ''
    # return '{:.1f}%'.format(pct) if pct >= 10 else ''

plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

```python
plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90, colors=colors, wedgeprops=wedgeprops, pctdistance=0.7)
plt.show()
```

```python
plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90, colors=colors, wedgeprops=wedgeprops, pctdistance=0.7, counterclock=False)
plt.show()
```

```python
plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90, colors=colors, wedgeprops=wedgeprops, pctdistance=0.7, counterclock=False, rotatelabels=45)
plt.show()
```
