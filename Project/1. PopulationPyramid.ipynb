# 1. Population Pyramid

## Population(M) in 2011

```python
import pandas as pd
df_m = pd.read_excel('201106_201106.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head()
```

```python
# Because of commas, we need to change data to integers
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int)
df_m
```

```python
df_m.iloc[0]
```

## Population(F) in 2011

```python
df_w = pd.read_excel('201106_201106.xlsx', skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w.head()
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns
```

```python
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
df_w.head()
```

## Data Visualization (2011)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, df_m.iloc[0] // 1000)
plt.barh(df_w.columns, -df_w.iloc[0] // 1000)

plt.title('Population of Korea in 2011')
plt.savefig('2011_PopulationPyramid.png', dpi=100)
plt.show()
```
