```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot(df.index, df['출생아 수'])
plt.plot(df.index, df['합계 출산율'])
```

```python
fig, ax1 = plt.subplots(figsize=(10, 7))
ax1.plot(df.index, df['출생아 수'])

ax2 = ax1.twinx() # both y values will have the same x-axis
ax2.plot(df.index, df['합계 출산율'])
```

```python
fig, ax1 = plt.subplots(figsize=(10, 7))
ax1.plot(df.index, df['출생아 수'], color='g', alpha=0.5)

ax2 = ax1.twinx()
ax2.plot(df.index, df['합계 출산율'], color='r', alpha=0.5)
```

```python
fig, ax1 = plt.subplots(figsize=(10, 7))
ax1.set_ylabel('Newborn Babies x 1000')
ax1.bar(df.index, df['출생아 수'], color='g', alpha=0.5)

ax2 = ax1.twinx()
ax2.set_ylabel('Total Fertility Rate')
ax2.plot(df.index, df['합계 출산율'], color='r', alpha=0.5)
```

```python
fig, ax1 = plt.subplots(figsize=(15, 5))
ax1.set_ylabel('Newborn Babies x 1000')
ax1.set_ylim(250, 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color='g', alpha=0.5)

ax2 = ax1.twinx()
ax2.set_ylabel('Total Fertility Rate')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color='r', alpha=0.5)
```

```python
fig, ax1 = plt.subplots(figsize=(15, 5))
ax1.set_ylabel('Newborn Babies x 1000')
ax1.set_ylim(250, 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color='g', alpha=0.5)
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 5, val, ha='center')

ax2 = ax1.twinx()
ax2.set_ylabel('Total Fertility Rate')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color='r', alpha=0.5)
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val + 0.05, val, ha='center')
```

```python
fig, ax1 = plt.subplots(figsize=(15, 5))
ax1.set_ylabel('Newborn Babies x 1000')
ax1.set_ylim(250, 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color='g')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 5, val, ha='center')

ax2 = ax1.twinx()
ax2.set_ylabel('Total Fertility Rate')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color='r', marker='o', ms=15, linewidth=5, markeredgecolor='w', markeredgewidth=3)
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val + 0.05, val, ha='center')
```

```python
fig, ax1 = plt.subplots(figsize=(15, 5))
fig.suptitle('Number of Newborn Babies and Fertility Rate')
ax1.set_ylabel('Newborn Babies x 1000')
ax1.set_ylim(250, 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color='g')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 7, val, ha='center')

ax2 = ax1.twinx()
ax2.set_ylabel('Total Fertility Rate')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['합계 출산율'], color='r', marker='o', ms=15, linewidth=5, markeredgecolor='w', markeredgewidth=3)
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val + 0.07, val, ha='center')
```
